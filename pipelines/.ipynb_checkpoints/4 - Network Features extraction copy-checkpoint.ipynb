```python
def average_nearest_neighbor_distance(coords):
    nbrs = NearestNeighbors(n_neighbors=2).fit(coords)
    distances, _ = nbrs.kneighbors(coords)
    return {'average_distance_NN':np.mean(distances[:, 1])}

avg_nn_dist = average_nearest_neighbor_distance(coordinates)
avg_nn_dist
```

```python
def centrality(edges):
    G = nx.Graph()
    G.add_edges_from(edges)
    degree_cent = nx.degree_centrality(G)
    closeness_cent = nx.closeness_centrality(G)
    DegC = {'Mean Centrality':pd.Series(degree_cent).mean(),
            'Median Centrality':pd.Series(degree_cent).median(),
            'Mean Closeness':pd.Series(closeness_cent).mean(),
            'Median Closeness':pd.Series(closeness_cent).median(),
            }
    return DegC

centrality = centrality(edges)
centrality
```

```python
labels = types
def shannon_diversity_index(labels):
    _, counts = np.unique(labels, return_counts=True)
    return { 'shannon_diversity_index': entropy(counts, base=2)}

shannon_index = shannon_diversity_index(labels)
shannon_index
```

```python
def simpson_diversity_index(labels):
    _, counts = np.unique(labels, return_counts=True)
    probabilities = counts / counts.sum()
    return {"simpson_diversity_index" : 1 - np.sum(probabilities**2)}

simpson_index = simpson_diversity_index(labels)
simpson_index
```

```python
def calculate_attribute_assortativity(edges, labels):
    G = nx.Graph()
    G.add_edges_from(edges)
    # Add attributes
    for node, label in enumerate(labels):
        G.nodes[node]['label'] = label
    # Attribute assortativity
    attribute_assortativity = nx.attribute_assortativity_coefficient(G, 'label')
    return {'Network_Assortativity':attribute_assortativity}

# Example data
attribute_assortativity = calculate_attribute_assortativity(edges, types)
attribute_assortativity
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Function to create a graph and add attributes to nodes
def create_graph_with_attributes(edges, labels):
    G = nx.Graph()
    G.add_edges_from(edges)
    # Assuming labels are provided in a list where the index corresponds to the node number
    for node, label in enumerate(labels):
        G.nodes[node]['type'] = label
    return G

# Function to compute the mixing matrix
def compute_mixing_matrix(G, attribute):
    # Using networkx to compute the attribute mixing matrix
    matrix = nx.attribute_mixing_matrix(G, attribute)
    return matrix

def compute_mixing_matrix_dataframe(G, attribute, normalized = True):
    # Compute the attribute mixing matrix
    matrix = nx.attribute_mixing_matrix(G, attribute, normalized=normalized)
    labels = sorted(set(nx.get_node_attributes(G, attribute).values()))

    # Transform the matrix to a DataFrame with three columns
    df = pd.DataFrame([(labels[i], labels[j], matrix[i, j])
                        for i in range(len(labels))
                        for j in range(len(labels))
                        if matrix[i, j] != 0],
                      columns=['a', 'b', 'value'])
    df['pairs'] = 'Mixing_Matrix_' + df['a'] + '|' + df['b']
    df.index = df['pairs']
    return df[['value']].to_dict()


# Visualization of the mixing matrix
def plot_mixing_matrix(matrix, labels):
    fig, ax = plt.subplots()
    cax = ax.matshow(matrix, cmap='Blues')
    fig.colorbar(cax)
    ax.set_xticklabels([''] + labels)
    ax.set_yticklabels([''] + labels)
    ax.set_xlabel('Node Attribute')
    ax.set_ylabel('Node Attribute')
    plt.title('Attribute Mixing Matrix')
    plt.show()




edges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 5)]
labels = ['Type1', 'Type2', 'Type1', 'Type3', 'Type2', 'Type3']


# Create graph and compute matrix
G = create_graph_with_attributes(edges, labels)
matrix = compute_mixing_matrix(G, 'type')
matrix

compute_mixing_matrix_dataframe(G, 'type')
# Unique labels sorted for consistent axis labeling
unique_labels = sorted(set(labels))
plot_mixing_matrix(matrix, unique_labels)
```

# Mixing matrix with mosna

### To mitigate the scale dependency, one approach is to normalize the mixing matrix by the product of the frequencies of each label type, rather than just by the total number of edges or by expected values that don't consider label distribution skew. This approach ensures that the matrix reflects the relative interaction rates more accurately and is not biased by label frequency.
###  

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# Create a graph and add attributes correctly to nodes
def create_graph_with_attributes(edges, labels):
    G = nx.Graph()
    G.add_edges_from(edges)
    # Add all nodes explicitly to ensure every node is included even if it has no edge
    for node, label in enumerate(labels):
        G.add_node(node, type=label)  # Make sure to use the correct attribute key
    return G

# Compute the normalized mixing matrix
def compute_normalized_mixing_matrix(G, attribute):
    # Retrieve node attributes
    labels = nx.get_node_attributes(G, attribute)
    unique_labels = sorted(set(labels.values()))
    label_index = {label: i for i, label in enumerate(unique_labels)}

    # Initialize raw mixing matrix
    raw_matrix = np.zeros((len(unique_labels), len(unique_labels)), dtype=int)
    for node1, node2 in G.edges():
        label1, label2 = labels[node1], labels[node2]
        index1, index2 = label_index[label1], label_index[label2]
        raw_matrix[index1][index2] += 1
        if index1 != index2:
            raw_matrix[index2][index1] += 1  # For undirected graph

    # Compute expected interactions
    label_counts = {label: 0 for label in unique_labels}
    for label in labels.values():
        label_counts[label] += 1
    total_nodes = sum(label_counts.values())

    expected_matrix = np.zeros_like(raw_matrix, dtype=float)
    for i, label1 in enumerate(unique_labels):
        for j, label2 in enumerate(unique_labels):
            expected_interactions = label_counts[label1] * label_counts[label2] / total_nodes
            expected_matrix[i][j] = expected_interactions

    # Normalize the mixing matrix
    with np.errstate(divide='ignore', invalid='ignore'):
        normalized_matrix = np.divide(raw_matrix, expected_matrix)
        normalized_matrix[~np.isfinite(normalized_matrix)] = 0

    return normalized_matrix, unique_labels

def plot_normalized_mixing_matrix(matrix, labels):
    fig, ax = plt.subplots()
    cax = ax.matshow(matrix, cmap='viridis')
    fig.colorbar(cax)
    plt.xticks(range(len(labels)), labels, rotation=90)
    plt.yticks(range(len(labels)), labels)
    ax.set_xlabel('Node Type')
    ax.set_ylabel('Node Type')
    plt.show()

# Example data


edges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 5)]
labels = ['Type1', 'Type2', 'Type1', 'Type3', 'Type2', 'Type3']


G = create_graph_with_attributes(edges, labels)


norm_matrix, labels = compute_normalized_mixing_matrix(G, 'type')
plot_normalized_mixing_matrix(norm_matrix, labels)
```

```python
def remove_duplicates_from_melted_df(melted_df):
    melted_df.columns = ['Row', 'Column', 'Score']
    melted_df['Sorted Row'] = melted_df.apply(lambda x: min(x['Row'], x['Column']), axis=1)
    melted_df['Sorted Column'] = melted_df.apply(lambda x: max(x['Row'], x['Column']), axis=1)
    # Drop original Row and Column to avoid confusion
    melted_df.drop(['Row', 'Column'], axis=1, inplace=True)
    melted_df = melted_df.drop_duplicates(subset=['Sorted Row', 'Sorted Column'])
    return(melted_df)
```

```python
def transform_score_matrix_to_vector(norm_matrix, labels):
    melted_df = pd.DataFrame(norm_matrix, columns=labels , index = labels).reset_index().melt(id_vars=['index'])
    df = remove_duplicates_from_melted_df(melted_df)
    df['pairs'] = 'Adjusted_Connectivity_' + df['Sorted Row'] + '_' + df['Sorted Column']
    df.index = df['pairs']
    return df[['Score']].to_dict()

transform_score_matrix_to_vector(norm_matrix, labels)
```

```python
def get_mosna_mixmat_vector(mixmat, attributes_list):
    mosna_mixing_mat = {}
    mixmat = pd.DataFrame(mixmat)
    mixmat.index = attributes_list
    mixmat.columns = attributes_list
    df_stacked = mixmat.stack().reset_index()
    df_filtered = df_stacked[df_stacked['level_0'] >= df_stacked['level_1']]

    for key, row in df_filtered.iterrows():
        id = 'MosnaMixingMatrix_' + row['level_0'] + "_" + row['level_1'] 
        mosna_mixing_mat[id] = row[0]

    return mosna_mixing_mat
```

# SquidPy

```python
import numpy as np
import pandas as pd
import squidpy as sq
import scanpy as sc

# Simulating data
np.random.seed(42)  # for reproducibility
num_cells = 100
coordinates = np.random.rand(num_cells, 2) * 100  # random coordinates in a 100x100 space

# Simulate cell phenotypes
phenotypes = np.random.choice(['Type1', 'Type2', 'Type3'], size=num_cells)
connectivity = np.random.randint(0, 2, (num_cells, num_cells))  # binary connectivity matrix

# Creating an AnnData object
adata = sc.AnnData(obs=pd.DataFrame(phenotypes, columns=['phenotype']), 
                   obsm={'spatial': coordinates})
adata.obsp['connectivities'] = connectivity

# Let's visualize the spatial distribution of cells
sc.pl.embedding(adata, basis='spatial', color='phenotype', title='Simulated Cell Types')
```

```python
SquidPy_Enrich_mat = {}

# Calculate neighborhood graph
sq.gr.spatial_neighbors(adata, coord_type='generic')  # You might adjust parameters based on your specific data setup

cluster_key='phenotype'
# Calculate neighborhood enrichment
sq.gr.nhood_enrichment(adata, cluster_key=cluster_key)
nes_key = f'{cluster_key}_nhood_enrichment'
cats = adata.obs[cluster_key].cat.categories
nes = pd.DataFrame(adata.uns[nes_key]['count'], index=cats, columns=cats)

df_stacked = nes.stack().reset_index()

df_filtered = df_stacked[df_stacked['level_0'] >= df_stacked['level_1']]

for key, row in df_filtered.iterrows():
    id = 'MosnaMixingMatrix_' + row['level_0'] + "_" + row['level_1'] 
    SquidPy_Enrich_mat[id] = row[0]

SquidPy_Enrich_mat
```
