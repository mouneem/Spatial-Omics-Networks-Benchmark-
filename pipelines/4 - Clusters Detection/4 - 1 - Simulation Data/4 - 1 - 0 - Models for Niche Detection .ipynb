# 1 - Isolated Niches (Different Cell Types): Spatial or Molecular

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to generate points within a circle
def generate_circle_points(center, radius, num_points, phenotype):
    angles = np.linspace(0, 2 * np.pi, num_points)
    radii = np.sqrt(np.random.rand(num_points)) * radius
    x = center[0] + radii * np.cos(angles)
    y = center[1] + radii * np.sin(angles)
    phenotypes = [phenotype] * num_points
    return pd.DataFrame({'x': x, 'y': y, 'phenotype': phenotypes})

# Parameters
center_type1 = (10, 10)
center_type2 = (25, 10)
radius = 5
num_points = 100

# Generate points for each circle
df_type1 = generate_circle_points(center_type1, radius, num_points, 'type1')
df_type2 = generate_circle_points(center_type1, radius, num_points, 'type2')
df_type3 = generate_circle_points(center_type1, radius, num_points, 'type3')

df_type4 = generate_circle_points(center_type2, radius, num_points, 'type4')
df_type5 = generate_circle_points(center_type2, radius, num_points, 'type5')
df_type6 = generate_circle_points(center_type2, radius, num_points, 'type6')

# Combine the data into a single DataFrame
nodes = pd.concat([df_type1, df_type2, df_type3, df_type4, df_type5, df_type6], ignore_index=True)

var = 'phenotype'
SpAn.plot_cells_with_circles(nodes[['x','y']], nodes[var], quantile_threshold=1, alpha=0.2)

cluster_type = 'Isolated'
nodes.to_csv(f'{output_dir}/4_1_{cluster_type}_Clusters/{cluster_type}_nodes.csv')
nodes[['x','y']].to_csv(f'{output_dir}/4_1_{cluster_type}_Clusters/{cluster_type}_coords.csv')
nodes[['phenotype']].to_csv(f'{output_dir}/4_1_{cluster_type}_Clusters/{cluster_type}_phenotype.csv')
pd.get_dummies(nodes[['phenotype']]).astype(int).to_csv(f'{output_dir}/4_1_{cluster_type}_Clusters/{cluster_type}_expression.csv')
```

# 4 - Gradual Transition Zones

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to generate points within a rectangle
def generate_rectangle_points(x_start, y_start, width, height, num_points, phenotype):
    x = np.random.uniform(x_start, x_start + width, num_points)
    y = np.random.uniform(y_start, y_start + height, num_points)
    phenotypes = [phenotype] * num_points
    return pd.DataFrame({'x': x, 'y': y, 'phenotype': phenotypes})

# Function to generate gradient phenotype distribution within a rectangle
def generate_gradient_rectangle(x_start, y_start, width, height, num_points, phenotype1, phenotype2):
    x = np.random.uniform(x_start, x_start + width, num_points)
    y = np.random.uniform(y_start, y_start + height, num_points)
    phenotypes = []

    for xi in x:
        # Calculate the proportion of phenotype1 vs phenotype2 based on the x position
        proportion_type1 = (x_start + width - xi) / width
        if np.random.rand() < proportion_type1:
            phenotypes.append(phenotype1)
        else:
            phenotypes.append(phenotype2)

    return pd.DataFrame({'x': x, 'y': y, 'phenotype': phenotypes})

# Parameters for the rectangles
width = 20
height = 10
num_points = 500

# Generate points for each rectangle
left_df = generate_rectangle_points(-width, 0, width, height, num_points, 'type1')
gradient_df = generate_gradient_rectangle(0, 0, width, height, num_points, 'type1', 'type2')
right_df = generate_rectangle_points(width, 0, width, height, num_points, 'type2')

# Combine the data into a single DataFrame
nodes = pd.concat([left_df, gradient_df, right_df], ignore_index=True)

# Plot using seaborn
plt.figure(figsize=(14, 8))
sns.scatterplot(data=nodes, x='x', y='y', hue='phenotype', palette='bright')
plt.axis('off')
plt.show()

cluster_type = 'Gradual_Transition'
nodes.to_csv(f'{output_dir}/4_4_{cluster_type}_Clusters/{cluster_type}_nodes.csv')
nodes[['x','y']].to_csv(f'{output_dir}/4_4_{cluster_type}_Clusters/{cluster_type}_coords.csv')
nodes[['phenotype']].to_csv(f'{output_dir}/4_4_{cluster_type}_Clusters/{cluster_type}_phenotype.csv')
pd.get_dummies(nodes[['phenotype']]).astype(int).to_csv(f'{output_dir}/4_4_{cluster_type}_Clusters/{cluster_type}_expression.csv')
```

# 5- Molecular Doublets (Niches = A+B | C+D)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to generate doublets with specific displacements
def generate_doublets(center, num_pairs, phenotype1, phenotype2):
    x1 = center[0] + np.random.uniform(-50, 50, num_pairs)
    y1 = center[1] + np.random.uniform(-50, 50, num_pairs)
    
    directions = np.random.choice(['right', 'left'], num_pairs)
    x2 = np.copy(x1)
    y2 = np.copy(y1)
    
    for i, direction in enumerate(directions):
        if direction == 'right':
            x2[i] += 1
        elif direction == 'left':
            x2[i] -= 1
        elif direction == 'top':
            y2[i] += 1
        elif direction == 'bottom':
            y2[i] -= 1
    
    data1 = {'x': x1, 'y': y1, 'phenotype': [phenotype1] * num_pairs}
    data2 = {'x': x2, 'y': y2, 'phenotype': [phenotype2] * num_pairs}
    
    df1 = pd.DataFrame(data1)
    df2 = pd.DataFrame(data2)
    
    return pd.concat([df1, df2], ignore_index=True)

# Parameters
num_pairs = 50

# Generate doublets for A+B and C+D
center_AB = (10, 10)
center_CD = (10, 10)

df_AB = generate_doublets(center_AB, num_pairs, 'A', 'B')
df_CD = generate_doublets(center_CD, num_pairs, 'C', 'D')

# Combine the data into a single DataFrame
nodes = pd.concat([df_AB, df_CD], ignore_index=True)

# Plot using seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=nodes, x='x', y='y', hue='phenotype', palette='bright')
plt.axis('off')
plt.show()

cluster_type = 'Doublets'
nodes.to_csv(f'{output_dir}/4_5_{cluster_type}_Clusters/{cluster_type}_nodes.csv')
nodes[['x','y']].to_csv(f'{output_dir}/4_5_{cluster_type}_Clusters/{cluster_type}_coords.csv')
nodes[['phenotype']].to_csv(f'{output_dir}/4_5_{cluster_type}_Clusters/{cluster_type}_phenotype.csv')
pd.get_dummies(nodes[['phenotype']]).astype(int).to_csv(f'{output_dir}/4_5_{cluster_type}_Clusters/{cluster_type}_expression.csv')
```

# 6 - Boundary Interaction Zones

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to generate points within a circle
def generate_circle_points(center, radius, num_points, phenotype):
    angles = np.random.uniform(0, 2 * np.pi, num_points)
    radii = radius * np.sqrt(np.random.uniform(0, 1, num_points))
    x = center[0] + radii * np.cos(angles)
    y = center[1] + radii * np.sin(angles)
    phenotypes = [phenotype] * num_points
    return pd.DataFrame({'x': x, 'y': y, 'phenotype': phenotypes})

# Function to generate points within a donut-shaped region
def generate_donut_points(center, inner_radius, outer_radius, num_points, phenotype):
    angles = np.random.uniform(0, 2 * np.pi, num_points)
    radii = np.sqrt(np.random.uniform(inner_radius**2, outer_radius**2, num_points))
    x = center[0] + radii * np.cos(angles)
    y = center[1] + radii * np.sin(angles)
    phenotypes = [phenotype] * num_points
    return pd.DataFrame({'x': x, 'y': y, 'phenotype': phenotypes})

# Parameters
num_points_inner = 200
num_points_outer = 200
center = (10, 10)
inner_radius = 3
outer_radius = 4

# Generate points for the inner circle with phenotypes A and B
df_inner_A = generate_circle_points(center, inner_radius, num_points_inner // 2, 'A')
df_inner_B = generate_circle_points(center, inner_radius, num_points_inner // 8, 'B')
df_inner_D = generate_circle_points(center, inner_radius, num_points_inner // 8, 'D')

# Generate points for the donut-shaped region with phenotypes C and D
df_outer_B = generate_donut_points(center, inner_radius, outer_radius, num_points_outer // 8, 'B')
df_outer_D = generate_donut_points(center, inner_radius, outer_radius, num_points_outer // 8, 'D')
df_outer_C = generate_donut_points(center, inner_radius, outer_radius, num_points_outer // 2, 'C')

# Combine the data into a single DataFrame
nodes = pd.concat([df_inner_A, df_inner_B, df_outer_C, df_outer_D], ignore_index=True)
SpAn.plot_cells_with_hulls(nodes[['x','y']], nodes[var], quantile_threshold=1, alpha=0.2)


cluster_type = 'Boundaries'
nodes.to_csv(f'{output_dir}/4_6_{cluster_type}_Clusters/{cluster_type}_nodes.csv')
nodes[['x','y']].to_csv(f'{output_dir}/4_6_{cluster_type}_Clusters/{cluster_type}_coords.csv')
nodes[['phenotype']].to_csv(f'{output_dir}/4_6_{cluster_type}_Clusters/{cluster_type}_phenotype.csv')
pd.get_dummies(nodes[['phenotype']]).astype(int).to_csv(f'{output_dir}/4_6_{cluster_type}_Clusters/{cluster_type}_expression.csv')
```

# 7 - Rare Cell Populations in Specific Microenvironments

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to generate points within a rectangle
def generate_rectangle_points(x_start, y_start, width, height, num_points, phenotype, exclusion_zone=None):
    x = []
    y = []
    while len(x) < num_points:
        x_rand = np.random.uniform(x_start, x_start + width)
        y_rand = np.random.uniform(y_start, y_start + height)
        if exclusion_zone:
            x_ex_start, y_ex_start, width_ex, height_ex = exclusion_zone
            if (x_ex_start <= x_rand <= x_ex_start + width_ex) and (y_ex_start <= y_rand <= y_ex_start + height_ex):
                continue
        x.append(x_rand)
        y.append(y_rand)
    phenotypes = [phenotype] * num_points
    return pd.DataFrame({'x': x, 'y': y, 'phenotype': phenotypes})

# Parameters for the rectangles
width_A = 20
height_A = 10
num_points_A = 200
num_points_B = int(0.1 * num_points_A)
num_points_C = int(0.05 * num_points_A)

width_B = width_A * 0.1
height_B = height_A * 0.1

width_C = width_A * 0.05
height_C = height_A * 0.05

# Generate a random location for rectangle B within rectangle A
x_start_B = np.random.uniform(0, width_A - width_B)
y_start_B = np.random.uniform(0, height_A - height_B)

x_start_C = np.random.uniform(0, width_A - width_C)
y_start_C = np.random.uniform(0, height_A - height_C)
x_start_C2 = np.random.uniform(0, width_A - width_C)
y_start_C2 = np.random.uniform(0, height_A - height_C)

# Generate points for each rectangle
exclusion_zone = (x_start_B, y_start_B, width_B, height_B)
df_A = generate_rectangle_points(0, 0, width_A, height_A, num_points_A, 'A', exclusion_zone)
df_B = generate_rectangle_points(x_start_B, y_start_B, width_B, height_B, num_points_B, 'B')
df_C = generate_rectangle_points(x_start_C, y_start_C, width_C, height_C, num_points_C, 'C')
df_C2 = generate_rectangle_points(x_start_C2, y_start_C2, width_C, height_C, num_points_C, 'C')

# Combine the data into a single DataFrame
nodes = pd.concat([df_A, df_B, df_C, df_C2], ignore_index=True)

SpAn.plot_cells_with_hulls(nodes[['x','y']], nodes[var], quantile_threshold=1, alpha=0.2, color_dict = {'A':'skyblue','B':'red','C':'purple'})

cluster_type = 'Rare_Populations'

nodes.to_csv(f'{output_dir}/4_7_{cluster_type}_Clusters/{cluster_type}_nodes.csv')
nodes[['x','y']].to_csv(f'{output_dir}/4_7_{cluster_type}_Clusters/{cluster_type}_coords.csv')
nodes[['phenotype']].to_csv(f'{output_dir}/4_7_{cluster_type}_Clusters/{cluster_type}_phenotype.csv')
pd.get_dummies(nodes[['phenotype']]).astype(int).to_csv(f'{output_dir}/4_7_{cluster_type}_Clusters/{cluster_type}_expression.csv')
```
